```
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os

path= '/content/drive/My Drive/IMAGEN/'
os.chdir(path)
#sirve ver el resultado de cada codigo en vivo
%matplotlib inline
```

```
def displaygray3(a, b, c, title1= "Original", title2 = "Editado", title3 = "Editado2"):
    plt.figure(figsize=(3,3))
    plt.subplot(131), plt.imshow(a, cmap='gray'), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(132), plt.imshow(b, cmap='gray'), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    plt.subplot(133), plt.imshow(c, cmap='gray'), plt.title(title3)
    plt.xticks([]), plt.yticks([])
    plt.show()
    
def displaygray2(a, b, title1= "Original", title2 = "Editado"):
    plt.figure(figsize=(10,10))
    plt.subplot(121), plt.imshow(a, cmap='gray'), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(122), plt.imshow(b, cmap='gray'), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    plt.show()
    
def displaycolor3(a, b, c, title1= "Original", title2 = "Editado", title3 = "Editado2"):
    plt.figure(figsize=(15,15))
    plt.subplot(131), plt.imshow(a), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(132), plt.imshow(b), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    plt.subplot(133), plt.imshow(c), plt.title(title3)
    plt.xticks([]), plt.yticks([])
    plt.show()
    
def displaycolor2(a, b, title1= "Original", title2 = "Editado"):
    plt.figure(figsize=(10,10))
    plt.subplot(121), plt.imshow(a), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(122), plt.imshow(b), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    plt.show()
    
def readRGB(imPath):
    img =cv2.imread(imPath)
    img =cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img
    
    
```

```
imgA = readRGB('hola.jpg')
imgM = readRGB('ico.jpg')
near_imgA = cv2.resize(imgA,None, fx=5, fy=5, interpolation = cv2.INTER_NEAREST)
near_imgB = cv2.resize(imgM,None, fx=5, fy=5, interpolation = cv2.INTER_NEAREST)
displaycolor2(near_imgA, near_imgB,'apple','señora')
```

```
imgA = readRGB('apple20x22.jpg')
imgM = readRGB('ico.jpg')
near_imgC = cv2.resize(imgA,None, fx=5, fy=5, interpolation = cv2.INTER_LINEAR)
near_imgD = cv2.resize(imgM,None, fx=5, fy=5, interpolation = cv2.INTER_LINEAR)
displaycolor2(near_imgC, near_imgD,'apple','señora')
```

```
imgA = readRGB('apple20x22.jpg')
imgM = readRGB('ico.jpg')
near_imgE = cv2.resize(imgA,None, fx=5, fy=5, interpolation = cv2.INTER_CUBIC)
near_imgF = cv2.resize(imgM,None, fx=5, fy=5, interpolation = cv2.INTER_CUBIC)
displaycolor2(near_imgE, near_imgF,'apple','señora')
```

```
methods = [None, 'none','nearest','bilinear','bicubic','spline16','spline36','hanning','hamming','hermite','kaiser','quadric','catrom','gaussian','bessel','mitchell','sinc','lanczos']

grid=imgA

fig, axes = plt.subplots(3,6,figsize=(40,10), subplot_kw={'xticks': [],'yticks' : []})

fig.subplots_adjust(hspace=0.3, wspace=0.05)

for ax, interp_method in zip(axes.flat,methods):
    ax.imshow(grid, interpolation=interp_method)
    ax.set_title(interp_method)
    
plt.show()
```

```
from PIL import Image
foto= Image.open('hola.jpg')
datos=list(foto.getdata())
datos_invert =[(255 -datos[x][0], 255 - datos[x][1], 255 - datos[x][2]) for x in range(len(datos))]
imagen_invert = Image.new('RGB', foto.size)
imagen_invert.putdata(datos_invert)

plt.imshow(np.asarray(imagen_invert))
```
